```python
import matplotlib
import matplotlib.pyplot as plt
import math
import random
class Punto():
    def __init__(self,la,al,La,Al):
        self.x=random.randint(la,La)
        self.y=random.randint(al,Al)

class Centro():
    def __init__(self,k,j):
        self.x=20*(k+1)
        self.y=20*(j+1)

def distancia(xc, yc, xp, yp):
    return math.sqrt((xc -xp)**2 + (yc - yp)**2)

def media(grupo):
    xm=0
    ym=0
    for i in grupo:
        xm+=i[0]
        ym+=i[1]
    if len(grupo)==0:
        return 0, 0
    else:
        return xm/len(grupo), ym/len(grupo)
    
if __name__=='__main__':
    universo=[]
    centros=[]
    for i in range(70):
        punto=Punto(10,10,40,40)
        universo.append([punto.x, punto.y])
    for i in range(50):
        punto=Punto(10,70,30,90)
        universo.append([punto.x, punto.y])
    for i in range(50):
        punto=Punto(65,10,90,35)
        universo.append([punto.x, punto.y])
    for i in range(150):
        punto=Punto(50,50,100,100)
        universo.append([punto.x, punto.y])
    
    x=[p[0] for p in universo]
    y=[p[1] for p in universo]
    plt.scatter(x,y)
    plt.show()
    
    for i in range(4):
        for j in range(4):
            centro=Centro(i,j)
            centros.append([centro.x, centro.y])
    
    for pasos in range(20):
        xc=[p[0] for p in centros]
        yc=[p[1] for p in centros]
        
        centros_m=[]
        for centro in centros:
            grupo=[]
            for punto in universo:
                if abs(punto[0]-centro[0])<=20 or abs(punto[1]-centro[1])<=20:
                    rango= round(distancia(centro[0], centro[1], punto[0], punto[1]),2)
                    if rango<=20:
                        grupo.append([punto[0], punto[1]])
            xm, ym=media(grupo)
            centros_m.append([xm, ym])
        centros=centros_m
        
    plt.scatter(xc, yc)
    plt.show()
```

```python
  # Sorting list of Integers in ascending 
centros.sort() 
media_c=[[]]
print(centros) 
```

```python
pos=0
i=0
for centro in centros:
    if i==0:
        actual=centro
    
    if distancia(centro[0], centro[1],actual[0],actual[1])<20:
        media_c[pos].append([centro[0], centro[1]])
    else:
        pos+=1
        i=-1
        media_c.append([[centro[0],centro[1]]])
    i+=1

print(media_c)
    
```

```python
centros=[]
for i in range(len(media_c)):
    centros.append(media(media_c[i]))
print(centros)
```

```python

grupos=[]
for i in range(4):
    grupos.append([])
    
for punto in universo:
    minimo=100
    ahora=0
    for centro in centros:
        variable=distancia(punto[0],punto[1], centro[0], centro[1])
        if variable<minimo:
            minimo=variable
            posicion=ahora
        ahora+=1
    grupos[posicion].append([punto[0], punto[1]])
        
for grupo in grupos:
    xg=[p[0] for p in grupo]
    yg=[p[1] for p in grupo]
    plt.scatter(xg,yg)
plt.show()
```
